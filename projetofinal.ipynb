```python
#pip install requests necessário
import pandas as pd 
import requests
import json
import sqlite3
```

```python
# Acessar API1 - IBGE UF 
url_uf = "https://servicodados.ibge.gov.br/api/v1/localidades/estados"
response_uf = requests.get(url_uf)
```

```python
# Função para verificar acesso à API e enviar uma notificação
# Parâmetro response: requisição http 
from plyer import notification

def notifica_status(response):
    if response.status_code == 200:
        data_json = response.json()
        notification.notify(
            title='API acessada com sucesso!',
            message='Prosseguir com extração.',
            app_name='Python',
            timeout=10
        )
    else:
        notification.notify(
            title='Erro ao acessar a API',
            message='Verifique se as informações estão corretas.',
            app_name='Python',
            timeout=10
        )
        
```

```python
notifica_status(response_uf)
```

```python
data_json = response.json()
data_json
```

```python
# Extrair dados API em tabela 
df_uf = pd.DataFrame(data_json)
df_uf
```

```python
# Criando função para verificar nulos no DataFrame
# Parâmetro df (DataFrame): o DataFrame que será verificado
def verificar_nulos(df):
     nulos = df.isnull().sum()
     if nulos.any():
        print("Há valores nulos nas seguintes colunas:")
        print(nulos[nulos > 0])
     else:
        print("Não há valores nulos no DataFrame.")
```

```python
verificar_nulos(df_uf)
```

```python
# Selecionar informações importantes da AP1 - IBGE UF
id = [x for x in df_uf['id']]
sigla = [x for x in df_uf['sigla']]
nome = [x for x in df_uf['nome']]
regiao = [x ['nome'] for x in df_uf['regiao']]
```

```python
# Transformar em tabela, alterando títulos
df_uf = pd.DataFrame({
    'Código':id,
    'Sigla':sigla,
    'Nome':nome,
    'Região':regiao,
})
df_uf.head()
```

```python
#Função para pesquisar no DataFrame com base no nome do território
#Parâmetros: df (DataFrame) nome (str)
def pesquisar_por_nome(df, nome):
    resultado = df[df['Nome'] == nome]
    if not resultado.empty:
        return resultado
    else:
        return "Nenhum dado encontrado para esse território."
```

```python
pesquisar_por_nome(df_uf,"Mato Grosso do Sul")
```

```python
## Função para exportar a tabela como um banco de dados
# Parâmetros df(DataFrame), nome_banco(str), nome_tabela(str)
def exportar_para_banco(df, nome_banco, nome_tabela):
    # Conecta ou cria o banco de dados
    conn = sqlite3.connect(nome_banco)
    # Exporta o DataFrame para tabela
    df.to_sql(nome_tabela, conn, if_exists='replace', index=False)
    # Fecha a conexão com o banco de dados
    conn.close()
    print(f"DataFrame exportado para o banco de dados '{nome_banco}' na tabela '{nome_tabela}'.")
```

```python
exportar_para_banco(df_uf, 'projetofinal.db', 'unidades_federativas')
```

```python
##Função para importar tabela
#Parâmetros nome_banco (str), nome_tabela(str)
def importar_tabela(nome_banco, nome_tabela):
    # Conectar ou criar o banco de dados
    conn = sqlite3.connect(nome_banco)
    # Executa uma consulta SELECT na tabela e converte em DataFrame
    df = pd.read_sql(f"SELECT * FROM {nome_tabela}", conn)
    # Fecha conexão
    conn.close()

    return df
```

```python
importar_tabela('projetofinal.db','unidades_federativas')
```

```python
##Função para exportar tabela em CSV
#Parâmetros nome_banco(str), nome_tabela(str), e arquivo_csv(str)
def exportar_tabela_csv(nome_banco, nome_tabela, arquivo_csv):
    conn = sqlite3.connect(nome_banco)
    # Executa uma consulta SELECT na tabela e converte em DataFrame
    df = pd.read_sql(f"SELECT * FROM {nome_tabela}", conn)
    # Fecha conexão
    conn.close()
    #Exporta o DataFrame para arquivo CSV
    df.to_csv(f"{arquivo_csv}.csv", index=False)

    print(f"Tabela '{nome_tabela}' exportada para o arquivo '{arquivo_csv}'.")
```

```python
exportar_tabela_csv('projetofinal.db','unidades_federativas','UF_siglas e codigos')
```

```python
## Função para verificar quantas tabelas existes no banco de dados
def listar_tabelas_bancodados(db_path):
    conn = sqlite3.connect(db_path)
    query = "SELECT name FROM sqlite_master WHERE type='table'"
    schema = pd.read_sql_query(query, conn)
    conn.close()
   
    return [schema]
```

```python
# Acessar API4 Censo Escolar
# A partir das orientações do IBGE, acessar resultados para matrículas da Educação Básica de Dourados MS - código 5003702
# Códigos obtidos do acesso à pesquisa: 77881 - Educação Infaltil | 5908 - Ensino Fundamental | 5913: Ensino Médio
url_4 = "https://servicodados.ibge.gov.br/api/v1/pesquisas/13/indicadores/77881|5908|5913/resultados/5003702"
response_4 = requests.get(url_4)
```

```python
notifica_status(response_4)
```

```python
data_json_4 = response_4.json()
data_json_4
```

```python
##Cria um dicionário vazio para armazenar os dados por ano

resultados = {}

## Itera sobre cada item nos dados
for item in data_json_4: 
    id_column = item['id'] 
    for res_item in item['res']: 
        for year, value in res_item['res'].items(): 
            if year not in resultados:
                resultados[year] = {} 
            resultados[year][id_column] = int(value) 
```

```python
#Exibe DataFrame
df_matriculasddos = pd.DataFrame(resultados)
df_matriculasddos
```

```python
# Transpondo linhas e colunas
df_matriculasddos = pd.DataFrame(resultados).T
df_matriculasddos
```

```python
#Troca título das colunas
df_matriculasddos.columns = ['Educação Infantil', 'Ensino Fundamental', 'Ensino Médio']

# Renomeia o índice (que são os anos) para 'Ano'
df_matriculasddos = df_matriculasddos.rename_axis('Ano').reset_index()

df_matriculasddos
```

```python
verificar_nulos(df_matriculasddos)
```

```python
# Adiciona uma nova coluna "Total" com a soma dos valores de cada linha
df_matriculasddos['Total'] = df_matriculasddos[['Educação Infantil', 'Ensino Fundamental', 'Ensino Médio']].sum(axis=1)

df_matriculasddos
```
